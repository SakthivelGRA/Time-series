```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm

# Load the dataset
df = pd.read_csv('Normalized+Age.csv')

# Display the first few rows of the dataset
print(df.head())

# Plot the Time Series
plt.figure(figsize=(10,6))
plt.plot(df['Year'], df['Age'], marker='o')
plt.title('Normalized Age over Years')
plt.xlabel('Year')
plt.ylabel('Normalized Age')
plt.grid(True)
plt.show()

# Perform basic statistical analysis
print(df.describe())

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns



# Plotting the time series
plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='Age', data=df, marker='o')

# Adding titles and labels
plt.title('Yearly Time Series of Normalized Average Age')
plt.xlabel('Year')
plt.ylabel('Normalized Average Age')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

# Load the dataset


# Perform the seasonal decomposition
result = seasonal_decompose(df['Age'], model='additive', period=1)

# Plot the original series and the trend component
plt.figure(figsize=(14, 7))

plt.subplot(211)
plt.plot(df['Year'], df['Age'], label='Original Series')
plt.plot(df['Year'], result.trend, label='Trend', linestyle='--')
plt.legend(loc='best')
plt.title('Original Series vs Trend')

plt.show()
```
